# ABCD Expert Rater Quiz Results

Thank you for doing the ABCD expert rater quiz! Below you will find the answers to the quiz and see how the other raters rated them. Please pay attention to any examples where your rating was very different from the rest of the raters.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import sys, os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import mode
from IPython.display import Image, Markdown, display
import warnings
warnings.simplefilter("ignore", FutureWarning)


all_raters_df = pd.read_csv("/Users/stevenmeisler/Documents/Penn/abcc-qsiprep/ge_philips_quiz_results/quiz_ratings_ge_philips.csv")
training_images_path = "/Users/stevenmeisler/Documents/Penn/abcc-qsiprep/expert_raters/training_images_ge_philips/"
ratername = os.getenv("RATERNAME")

rater_df = all_raters_df[all_raters_df["rater"]==ratername]

def grade_rater(subject_id, real_grade):
    sub_ratings_df = all_raters_df[all_raters_df['subject'] == subject_id]
    rating_lookup = {
        -2: "Definitely Fail",
        -1: "Probably Fail",
        0: "Not Sure",
        1: "Probably Pass",
        2: "Definitely Pass"
    }

    sub_ratings_df = all_raters_df[all_raters_df['subject'] == subject_id]
    group_mode = int(mode(sub_ratings_df["rating"]).mode)
    group_decision_numeric = sub_ratings_df["rating"].mean()
    group_decision_text = "Pass" if group_decision_numeric > 0 else "Fail"
    group_rating_text = rating_lookup[group_mode]

    your_rating = int(rater_df[rater_df["subject"]==subject_id]["rating"])
    your_rating_text = rating_lookup[your_rating]
    your_decision_text = "Pass" if your_rating > 0 else "Fail"

    real_descision_text = "Pass" if real_grade > 0 else "Fail"


    # What color should the section title be?
    if your_rating == 0:
        ans_color = "darkblue"
        your_decision_text = "Not Sure"
        your_score = "You weren't sure"
    #elif your_decision_text == group_decision_text:
    elif your_decision_text == real_descision_text:
        ans_color = "black"
        your_score = "You were correct!"
    else:
        ans_color = "maroon"
        your_score = "You missed it :("


    group_decision_paragraph = f"""
 * The group of raters decided that this image overall
   would be considered a "{group_decision_text}". 
 * The most common rating by the raters was "{group_rating_text}".
    """

    # Did you guess with the group?
    real_score_paragraph = f"The quiz answer key is {real_grade}, or {rating_lookup[real_grade]}."

    if your_decision_text == group_decision_text:
        your_score_paragraph = f'You, like the group decided that this image is a "{group_decision_text}"'

        if your_rating_text == group_rating_text:
            your_score_paragraph += f" and were in exact agreement with most raters."
        else:
            your_score_paragraph += f', but you selected "{your_rating_text}" as your rating.'

    else:
        if your_rating == 0:
            your_score_paragraph = "You were not sure how to rate this image. *It might be a good idea to re-examine this subject.*"
        else: 
            your_score_paragraph = "Your rating was inconsistent with the group. *It might be a good idea to re-examine this subject.*"

        if group_mode in (-1, 0, 1):
            your_score_paragraph += " Most other raters were somewhat uncertain about this image."
        else:
            your_score_paragraph += f" Most other raters were sure that the image was a {group_decision_text}."

    display(Markdown(f"""##  <font color='{ans_color}'>{subject_id}: {your_score}</font>
                     
{real_score_paragraph}                     

{group_decision_paragraph}

{your_score_paragraph}

    """))

    # Plot the figure
    fig, ax = plt.subplots(figsize=(6,2))
    ax.hist(sub_ratings_df["rating"], bins=np.arange(6)-2.5, edgecolor="black")
    ax.set_xticks([-2, -1, 0, 1, 2])
    ax.set_xticklabels(
        ["Definitely\nFail", "Probably\nFail", "Not Sure", 
        "Probably\nPass", "Definitely\nPass"])

    ax.axvline(your_rating, lw=5, color=ans_color)
    ax.set_title(f"Ratings (yours is a {ans_color} line)");
```
